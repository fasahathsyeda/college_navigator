```python
geocoder = OpenCageGeocode(api_key)
```

```python
col_data=pd.read_csv("../Data/ForbesAmericasTopColleges2019.csv")
col_data.head()
```

```python
list_lat = []   # create empty lists

list_long = []



for index, row in col_data.iterrows(): # iterate over rows in dataframe



    City = row['City']
    State = row['State']       
    query = str(City)+','+str(State)

    results = geocoder.geocode(query)   
    lat = results[0]['geometry']['lat']
    long = results[0]['geometry']['lng']

    list_lat.append(lat)
    list_long.append(long)


# create new columns from lists    

col_data['lat'] = list_lat   

col_data['lon'] = list_long
```
