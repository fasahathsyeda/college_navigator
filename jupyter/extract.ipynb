```python
import pandas as pd
```

```python
#reading data in df
college_data1=pd.read_excel("../Data/Colleges_and_universities_usa.xlsx")
college_data1.head()
```

```python
# converting data to upper case
college_data1_df=college_data1.apply(lambda x: x.astype(str).str.upper())
```

```python
college_data1_df.head()
```

```python
# converting column name to upper case
college_data1_df.columns= college_data1.columns.str.upper()
```

```python
college_data1_df.columns
```

```python
# renaming the columns
college_data1_df.rename(columns={"SCHOOL NAME":"NAME","ZIPCODE":"ZIP"}, inplace=True)
college_data1_df.columns
```

```python
# redaing second data file
college_data2=pd.read_csv("../Data/Colleges_and_Universities.csv")
college_data2.head()
```

```python
# leaving unwanted columns
cols=['NAME','ADDRESS','CITY','STATE','ZIP','TELEPHONE','POPULATION','LATITUDE','LONGITUDE','WEBSITE']
college_data2_df=college_data2[cols].copy()
college_data2_df.head()
```

```python
college_data1_df.isnull().sum()
```

```python
college_data1_df.dtypes
```

```python
college_data2_df.dtypes
```

```python
#college_data2.astype({'ZIP': str}).dtypes
# converting Zip column to string type
college_data2_df['ZIP'] = college_data2_df['ZIP'].astype(str)
```

```python
college_data2_df.dtypes
```

```python
# merging the above data frames
college_data=pd.merge(college_data1_df,college_data2_df,how='inner', on=['NAME','CITY'])
college_data.columns
```

```python
college_data.head(20)
```

```python
college_data.count()
```

```python
# taking only subset of columns required and renaming them
cols=['NAME','ADDRESS_x','CITY','STATE_y','ZIP_x','TYPE',
       'DEGREES OFFERED', 'CAMPUS SETTING', 'CAMPUS HOUSING',
       'STUDENT POPULATION', 'UNDERGRADUATE STUDENTS', 'PUBLIC/PRIVATE',
       'PUBLISHED IN-STATE TUITION AND FEES ($)',
       'PUBLISHED OUT-OF-STATE TUITION AND FEES ($)', 'TELEPHONE', 'POPULATION', 'LATITUDE', 'LONGITUDE', 'WEBSITE']
college_data_df=college_data[cols].copy()
```

```python
college_data_df.head()
college_data_df.count()
```

```python
college_data_df.rename(columns={"ADDRESS_x":"ADDRESS","STATE_y":"STATE","ZIP_x":"ZIP"}, inplace=True)
college_data_df.head()
```

```python
# reading the top universities data
top_colleges=pd.read_csv("../Data/ranking.csv")
top_colleges.head()
```

```python
# Converting data and columns name to upper case
top_colleges=top_colleges.apply(lambda x: x.astype(str).str.upper())
top_colleges.columns= top_colleges.columns.str.upper()
top_colleges.head()
```

```python
# merging above dataframe with the previously mearged df
college_data_final=pd.merge(college_data_df,top_colleges,how='left', on=['NAME'])
college_data_final.dtypes
# converting RANK column dtype to float from str
college_data_final['RANK'] = college_data_final['RANK'].astype(float)
```

```python
college_data_final.dtypes
```

```python
college_data_final.sort_values(by=['RANK'], inplace=True)
```

```python
# deleting student_population_y
del college_data_final['STUDENT POPULATION_y']
```

```python
college_data_final.head()
```

```python
college_data_final.rename(columns={"STUDENT POPULATION_x":"STUDENT POPULATION"}, inplace=True)
college_data_final.head()
```

```python
# transfering the df to csv
college_data_final.to_csv("../Data/Final_College_Data.csv",index=False)
```

```python
import json
json_data=[]
data=[]
dict_json={}
for index,row in college_data_final.iterrows():
    geometry={
        'x':row['LATITUDE'],
           'y':row['LONGITUDE']
       }
    properties={
       'NAME':row['NAME'],
       'ADDRESS':row['ADDRESS'],
        'CITY':row['CITY'],
        'STATE':row['STATE'],
        'ZIP':row['ZIP'],
        'RANK':row['RANK'],
        'TYPE':row['TYPE'],
        'DEGREES OFFERED':row['DEGREES OFFERED'],
        'CAMPUS SETTING':row['CAMPUS SETTING'],
        'CAMPUS HOUSING':row['CAMPUS HOUSING'], 
        'STUDENT POPULATION':row['STUDENT POPULATION'],
        'UNDERGRADUATE STUDENTS':row['UNDERGRADUATE STUDENTS'], 
        'PUBLIC/PRIVATE':row['PUBLIC/PRIVATE'],
        'PUBLISHED IN-STATE TUITION AND FEES ($)':row['PUBLISHED IN-STATE TUITION AND FEES ($)'],
        'PUBLISHED OUT-OF-STATE TUITION AND FEES ($)':row['PUBLISHED OUT-OF-STATE TUITION AND FEES ($)'], 
        'TELEPHONE':row['TELEPHONE'],
        'POPULATION':row['POPULATION'], 
        'LATITUDE':row['LATITUDE'], 
        'LONGITUDE':row['LONGITUDE'], 
        'NET PRICE':row['NET PRICE'], 
        'AVERAGE GRANT AID':row['AVERAGE GRANT AID'],
        'TOTAL ANNUAL COST':row['TOTAL ANNUAL COST'], 
        'ALUMNI SALARY':row['ALUMNI SALARY'], 
        'ACCEPTANCE RATE':row['ACCEPTANCE RATE'],
        'WEBSITE':row['WEBSITE']
        }
    dict_json[index]={
        'geometry':geometry,
        'properties':properties
    }

    json_data.append(dict_json[index])
print(dict_json[0])
        
                
    
```

```python
with open("../Data/Final_College_Data.json", 'w') as outfile:
    outfile.write(json.dumps(json_data))
```
